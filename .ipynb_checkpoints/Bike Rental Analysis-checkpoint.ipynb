# Importing Libraries and Data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pyplot
```

```python
data = pd.read_csv("cyclistic_data.csv", delimiter = ';')
```

# Basic EDA

```python
data.shape
```

```python
data.info()
```

```python
data.head()
```

# Data Cleaning

The dataset is currently quite large. I will decrease the number of columns to only relevant features.

This will help us understand customer trends a little better.

```python
data = data[['rideable_type', 'member_casual', 'ride_time', 'day_of_week']]
```

```python
data.isna().value_counts()
```

```python
data.head()
```

I'll need to inspect the distribution of the numerical data

```python
data['ride_time'].plot.box(color='black')

pyplot.show()
```

There are some serious outliers in the data

I'll set upper and lower quantiles and filter the data

```python
q_low = data['ride_time'].quantile(0.01)
q_high = data['ride_time'].quantile(0.99)

data = data[(data['ride_time'] < q_high) & (data['ride_time'] > q_low)]
```

```python
data['ride_time'].plot.box(color = 'black')

pyplot.show()
```

```python
data['ride_time'].mean()
```

The average ride time for all customers in 14.27 minutes

# Analyzing Categorical Features

```python
categorical_features = data[['rideable_type', 'member_casual', 'day_of_week']]
```

```python
figure, axis = pyplot.subplots(1, 3,
                                figsize = (10, 5))

for feature, subplot in zip(categorical_features,
                            axis.flatten()):

          data[feature].value_counts().plot(kind='bar',
                                          color = 'black', edgecolor='yellow',
                                          ax = subplot,
                                          title = feature)
```

Here we see the total number of rides for each of the categories

### Let's look at days of the week

```python
data.groupby('day_of_week')['ride_time'].mean().sort_values(ascending=False)
```

People ride the longest, on average, on Sundays

```python
data['day_of_week'].value_counts().sort_values().plot(color = 'black')
pyplot.ylabel('Total Number of Rides')
pyplot.xlabel('Days of the Week')

pyplot.show()
```

But just like we saw in the bar charts above, Sunday is one of the least popular days for riding. While Saturday is the most popular.

```python
data.groupby('day_of_week')['ride_time'].sum()
```

So the company sells the most ride time on Friday

Now we'll see if any other features affect ride time

### Membership Status

```python
data.groupby('member_casual')['ride_time'].sum()
```

Casual members take less rides, but they spend more time riding

# Correlations

We'll use one hot encoding to translate our categorical variables so that we can run a correlation on them

```python
ohe = pd.get_dummies(data, columns = ['member_casual', 'day_of_week', 'rideable_type'])
```

```python
ohe.corr()['ride_time']
```

We can see that there is a slight correlation between casual members and ride time. As well as a slight correlation between docked bikes and ride time

# Conclusion

```python
average_ride_time = data.groupby('day_of_week')['ride_time'].mean()

riders_count = data['day_of_week'].value_counts()

revenue_by_user_type = data.groupby('member_casual')['ride_time'].sum()

fig, axes = pyplot.subplots(2, 2, figsize=(14, 10))

axes[0, 0].bar(average_ride_time.index, average_ride_time, color = 'black', edgecolor='yellow')
axes[0, 0].set_xlabel('Day of Week')
axes[0, 0].set_ylabel('Average Ride Time')
axes[0, 0].set_title('Average Ride Time by Day of Week')

axes[0, 1].bar(riders_count.index, riders_count, color = 'black', edgecolor='yellow')
axes[0, 1].set_xlabel('Day of Week')
axes[0, 1].set_ylabel('Number of Riders')
axes[0, 1].set_title('Number of Riders by Day of Week')

axes[1, 0].pie(revenue_by_user_type, labels=revenue_by_user_type.index, autopct='%1.1f%%', colors = ['black', 'yellow'])
axes[1, 0].set_title('Revenue Comparison: Casual vs. Member Users')


fig.delaxes(axes[1, 1])
fig.tight_layout()

pyplot.show()
```

From these observations we can make a few determinations
 - Average ride time is highest on Sundays
 - There are the most riders on Fridays
 - Casual members are the most profitable users
